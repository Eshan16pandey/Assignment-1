```python
import pandas as pd

import numpy as np
```

```python
Mean_Points = df['Points'].mean()
Median_Points = df['Points'].median()
Mode_Points = df['Points'].mode().tolist()
Variance_Points = df['Points'].var()
Std_dev_Points = df['Points'].std()
Range_points = df['Points'].max() - df['Points'].min()

print("Mean of Points:",Mean_Points)
print("Median of Points:",Median_Points)
print("Mode of Points:",Mode_Points)
print("Variance of Points:",Variance_Points)
print("Standard deviation of Points:",Std_dev_Points)
print("Range for Points:",Range_points)






```
